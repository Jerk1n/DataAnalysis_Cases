```python
## 04 Numpy 随机数

numpy.random 能生成多种概率分布的随机样本，是后期数据分析中常用的工具
```

```python
# 生成随机数
import numpy as np

samples = np.random.normal(size=(4,4))  # 生成标准正太分布 4*4
print(samples)
```

```python
# np.random.rand()  生成一个[0,1) 之间的随机浮点数或者N维浮点数组——均匀分布

a = np.random.rand()
print(a, type(a))

b = np.random.rand(4)
print(b, type(b))

c = np.random.rand(2,3)
print(c, type(c))

import matplotlib.pyplot as plt
%matplotlib inline

s1 = np.random.rand(1000)
s2 = np.random.rand(1000)
plt.scatter(s1,s2)
```

```python
# np.random.randn()  生成一个浮点数或者N维浮点数组——正太分布

s1 = np.random.randn(1000)
s2 = np.random.randn(1000)
plt.scatter(s1,s2)
```

```python
# np.random.randint(low, high=None, size=None, dtype='1')  生成一个随机整数或N维整数数组
# size 控制形状，low 和 high 控制取值范围，如果没定义 high,取值范围[0,low)，否则取值范围[low,high)

print(np.random.randint(3))   # low=3，生成1个 [0,3) 之间随机整数

print(np.random.randint(3, size=5))   # low=3，生成5个 [0,3) 之间随机整数

print(np.random.randint(3, 6, size=5))   # low=3，high=6，生成5个 [3,6) 之间随机整数

print(np.random.randint(3, size=(2,3)))   # low=3，size=(2,3)，生成一个值为[0,3)的2*3形状整数数组

print(np.random.randint(3, 6, size=(2,3)))   # low=3，high=6，生成一个值为[3,6)的2*3形状整数数组
```
