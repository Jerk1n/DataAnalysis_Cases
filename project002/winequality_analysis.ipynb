## 数据集包含下列字段

- fixed acidity：固定酸度
- volatile acidity：挥发性酸度
- citric acid：柠檬酸
- residual sugar：残糖
- chlorides：氯化物
- free sulfur dioxide：游离二氧化硫	
- total sulfur dioxide：总二氧化硫
- density：密度
- pH：PH 值
- sulphates：硫酸盐	
- alcohol：酒精度
- quality：质量，0 到 10 之间的得分（葡萄酒专家至少 3 次评估的中值）

# 重新整理数据集

## 合并两个文件

```python
import pandas as pd

df1 = pd.read_csv('winequality-red.csv',sep=';')
df2 = pd.read_csv('winequality-white.csv',sep=';')
# 红、白葡萄酒分别新增一列类型，填充为 red 和 white
df1['type'] = 'red'
df2['type'] = 'white'
df = pd.concat([df1,df2],ignore_index=True)
df.to_csv('winequality_edited.csv')
```

```python
wine_df = pd.read_csv('winequality_edited.csv',index_col=0)
wine_df.head()
```

### 答：酒精度含量高的，是否质量也更高？

是的

```python
print(wine_df['alcohol'].median())

wine_df_low = wine_df.query('alcohol <= 10.3')
wine_df_high = wine_df.query('alcohol > 10.3')

print(wine_df_low['quality'].mean())
print(wine_df_high['quality'].mean())
```

### 答：残糖含量越高，质量是否越高？

```python
print(wine_df['residual sugar'].median())

wine_df_1 = wine_df.query('residual sugar <= 3.0')
wine_df_2 = wine_df.query('residual sugar > 3.0')

print(wine_df_1['quality'].mean())
print(wine_df_2['quality'].mean())
```

## 用数据可视化看数据趋势

### 答：哪种葡萄酒的质量更高？

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

wine_df = pd.read_csv('winequality_edited.csv', index_col=0)
wine_df.head()
```

```python
color_means = wine_df.groupby('type')['quality'].mean()
color_means.plot(kind='bar', title='Average Wine Quality', color=['red','blue'], alpha=.7)
plt.xlabel('Colors', fontsize=18)
plt.ylabel('Quality', fontsize=18)
```

```python
counts = wine_df.groupby(['quality','type']).count()['pH']
counts.plot(kind='bar', title='Counts by Wine Color and Quality',color=['red','blue'], alpha=.7)
plt.xlabel('Quality and Color', fontsize=18)
plt.ylabel('Count', fontsize=18)
```

```python
totals = wine_df.groupby('type').count()['pH']
proportions = counts / totals
proportions.plot(kind='bar', title='Proportion by Wine Color and Quality',color=['red','blue'], alpha=.7)
plt.xlabel('Quality and Color', fontsize=18)
plt.ylabel('Proportion', fontsize=18)
```
