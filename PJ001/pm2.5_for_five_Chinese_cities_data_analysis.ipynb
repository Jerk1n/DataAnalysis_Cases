## 一、观察数据

数据集包括 5 份 csv 文件，是中国五个城市（北京、上海、成都、广州、沈阳）2010 年 1 月 1 日至 2015 年 12 月 31 日的空气和气象数据，数据中有大量缺失值 NaN, 数据列名包括：

- No: 行号
- year: 年份
- month: 月份
- day: 日期
- hour: 小时
- season: 季节
- PM_: pm2.5的浓度（ug/m^3）
- DEWP: 露点（摄氏温度）指在固定气压之下，空气中所含的气态水达到饱和而凝结成液态水所需将至的温度。
- TEMP: Temperature（摄氏温度）
- HUMI: 湿度（%）
- PRES: 气压（hPa）
- cbwd: 组合风向
- lws: 累计风速（m/s）
- precipitation: 降水量/时（mm）
- iprec: 累计降水量（mm）
